```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
import time
import psutil
import numpy as np
from sklearn import datasets

file_path = 'D:\Programming\Datasets\Classification\Stroke\stroke.csv'
    
df = pd.read_csv(file_path)
```

```python
def replace_with_noise(dataframe, percentage=0.1):
    # Setting the seed to ensure reproducibility
    np.random.seed(0)

    # Generate the random indices to be replaced
    num_rows = len(dataframe)
    num_replace = int(num_rows * percentage)
    idx = np.random.choice(np.arange(num_rows), size=num_replace, replace=False)

    # Generate noise equivalent to the number of replacements
    noise = np.random.normal(0, 1, num_replace)

    # Make a copy of the dataframe
    df_noisy = dataframe.copy()

    # Replace the selected indices with noise
    last_col = df_noisy.columns[-1]
    
    for column in df_noisy.columns[:-1]:
        df_noisy.loc[idx, column] = noise
        
    df_noisy.loc[idx, last_col] = df_noisy.loc[idx, last_col]

    return df_noisy
```

```python
df_noisy1 = replace_with_noise(df, 0.01)
df_noisy1
```

```python
features = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level']
target = 'stroke'

X = df_noisy1[features]
y = df_noisy1[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
svm_classifier = SVC(kernel='linear')

start_training_time = time.time()

svm_classifier.fit(X_train, y_train)

end_training_time = time.time()
training_time_duration = end_training_time - start_training_time
```

```python
start_testing_time = time.perf_counter()

# Predict on the test set
y_pred = svm_classifier.predict(X_test)

end_testing_time = time.perf_counter()
testing_time_duration = end_testing_time - start_testing_time
```

```python
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

print(f"Training Time: {training_time_duration:.8f} seconds")
print(f"Testing Time: {testing_time_duration:.8f} seconds")
```
