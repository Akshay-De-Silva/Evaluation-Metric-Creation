```python
import tensorflow as tf
from tensorflow.keras import layers, models
import time
import psutil  # Import the psutil library
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Create a simple neural network model
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(1)  # Output layer for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=[])

# Load a dummy dataset (MNIST)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Training loop with CPU usage tracking and RMSE, R^2 calculation
num_epochs = 5
check_interval = 1
total_cpu_usage = 0  # To accumulate total CPU usage
total_rmse = 0  # To accumulate total RMSE
total_r2 = 0  # To accumulate total R^2
num_checks = 0  # To count the number of checks

for epoch in range(num_epochs):
    # Train the model
    start_time = time.time()

    # Track CPU usage during training
    cpu_percent = psutil.cpu_percent(interval=1)  # Measure CPU usage for the last 1 second
    model.fit(x_train, y_train, epochs=1, verbose=2)

    # Predict on the test set
    y_pred = model.predict(x_test)
    y_pred = np.squeeze(y_pred)  # Remove singleton dimensions
    y_test_flat = y_test.flatten()

    # Calculate RMSE and R^2
    rmse = np.sqrt(mean_squared_error(y_test_flat, y_pred))
    r2 = r2_score(y_test_flat, y_pred)

    end_time = time.time()

    # Accumulate CPU usage, RMSE, and R^2
    total_cpu_usage += cpu_percent
    total_rmse += rmse
    total_r2 += r2
    num_checks += 1

    # Print information
    print(f"Epoch {epoch + 1}/{num_epochs} - CPU Usage: {cpu_percent}%, RMSE: {rmse:.4f}, R^2: {r2:.4f}")

    # Optional: Add a delay to avoid overwhelming the CPU monitoring tool
    time.sleep(5)

# Calculate and print average CPU usage, RMSE, and R^2
average_cpu_usage = total_cpu_usage / num_checks
average_rmse = total_rmse / num_checks
average_r2 = total_r2 / num_checks
print(f"Average CPU Usage: {average_cpu_usage}%, Average RMSE: {round(average_rmse, 4)}, Average R^2: {round(average_r2, 4)}")
```
